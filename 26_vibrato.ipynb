```python

import math
import numpy as np
import scipy.io.wavfile
import librosa
import matplotlib.pyplot as plt
```

```python
def generate_sinusoid(sampling_rate, frequency, duration):
  sampling_interval = 1.0 / sampling_rate
  t = np.arange(sampling_rate * duration) * sampling_interval
  waveform = np.sin(2.0 * math.pi * frequency * t)
  return waveform

def vibrato(waveform, sampling_rate: int, frequency: float, depth: float, tau: float):
  vibrato_waveform = generate_sinusoid(sampling_rate, frequency, len(waveform) / sampling_rate)
  changed = waveform.copy()
  for i in range(len(waveform)):
    taui = int(tau + depth * vibrato_waveform[i])
    if taui < 0:
      taui = 0
    elif taui >= len(waveform):
      taui = len(waveform) - 1
    changed[i] = waveform[i - taui]
  return changed

def export_wav(filename, sampling_rate, waveform):
  waveform = np.int16(waveform * 32767.0)
  scipy.io.wavfile.write(filename, sampling_rate, waveform)

SR = 16000
filename = 'aiueo-long'
x, _ = librosa.load(f'audio/{filename}.wav', sr=SR)
FREQ = 10
SIZE_FRAME = 2048
SHIFT_SIZE = 16000 / 100  # 10 msec
DEPTH = 10
TAU = 10
x_changed = vibrato(x, SR, FREQ, DEPTH, TAU)
fig = plt.figure()
fig.savefig(f'plot/vibrato/{filename}-{FREQ}-{int(DEPTH * 10)}.png')
export_wav(f'audio/vibrato/{filename}-{FREQ}-{int(DEPTH * 10)}.wav', SR, x_changed)
```
